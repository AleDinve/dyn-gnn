```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 18})
```

```python
seq_len_list = [4,5]
for seq_len in seq_len_list:
    data = pd.read_csv('dynamics_16_10_T'+str(seq_len)+'.csv')
    data = data[data['Number of timestamps']==seq_len]
    data = data[data['GNN hidden dimension']<9]
    plot = sns.lineplot(data=data, x='Epoch', y='train accuracy', hue='GNN hidden dimension',palette=sns.color_palette())
    plt.legend(title='hidden feature size')
    plt.show()
    plot.figure.savefig('pics/acc_timelength_'+str(seq_len)+'_16_10_1.pdf',bbox_inches='tight' )
    
    # plt.close()
```
